```python
import pandas as pd
import numpy as np
```

```python
# Show all rows and columns without truncation
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
# Importing dataframes
bacteria_path = 'https://raw.githubusercontent.com/HackBio-Internship/2025_project_collection/refs/heads/main/bacteria.csv'
bacteria_data = pd.read_csv(bacteria_path)
bacteria_data.head(5)
```

```python
# Part 1: Basic Data Inspection
# Shape of the dataset
print('Shape:') 
bacteria_data.shape
```

```python
# List of all column names
print('Columns:')
bacteria_data.columns.tolist()
```

```python
# Part 2: Data Selection
# Select row 15
print('Row 15:')
bacteria_data.iloc[15]
```

```python
# Select the column species
print('Column species:')
bacteria_data[['species']]
```

```python
# Retrieve the value at row 15 for the column Isolation origin
print('Value at row 15, Isolation origin:')
bacteria_data.loc[15, 'Isolation origin']
```

```python
# Part 3: Data Filtering
# Show only rows (genes) where carb_fit is greater than 0.5
print('Rows where carb_fit > 0.5:')
bacteria_data[bacteria_data['carb_fit'] > 0.5]
```

```python
# Part 4: Handling Missing Data
# Create a copy of bacteria_data as bacteria_data_missing
bacteria_data_missing = bacteria_data.copy()

# Count the number of missing values per row
print('Missing values per row:')
bacteria_data_missing.isna().sum(axis=1)
```

```python
# Create a new version of the dataset where missing values are filled with 1
print('Dataset with missing values filled with 1:')
bacteria_data_missing.fillna(1)
```

```python
# Create another version where columns containing missing values are dropped entirely
print('Dataset with columns containing missing values dropped:')
bacteria_data_missing.dropna()
```

```python
# Part 5: Checking Missing Data in Main Dataset
# Check whether missing values exist
print('Missing values per column in bacteria_data:')
bacteria_data_missing.isna().sum()
```

```python
# If yes, report how many missing values are present (total or per column)
print('Total missing values:', bacteria_data_missing.isna().sum().sum())
```

```python
# Part 6: Column Renaming
# Rename C1 to UMAP1 and C2 to UMAP2
bacteria_data.rename(columns={'C1': 'UMAP1', 'C2': 'UMAP2'}, inplace=True)
print('Columns after renaming:')
bacteria_data.columns.tolist()
```

```python
# Confirm as DataFrame
print('Confirming DataFrame:')
bacteria_data.head()
```

```python
# Part 7: Sorting
# By the column BSL
print('Data sorted by BSL:')
bacteria_data.sort_values(by='BSL')
```

```python
# In descending order (highest to lowest)
print('Data sorted by BSL (descending order):')
bacteria_data.sort_values(by='BSL', ascending=False)
```

```python
# Part 8: Groups Analysis
# the mean of cipro_fit grouped by label type
print('Mean of cipro_fit by labels:')
bacteria_data.groupby('labels')['cipro_fit'].mean()
```

```python
# the mean of carb_fit grouped by label type
print('Mean of carb_fit by labels:')
bacteria_data.groupby('labels')['carb_fit'].mean()
```

```python
# the maximum of cipro_fit grouped by label type
print('Max of cipro_fit by labels:')
bacteria_data.groupby('labels')['cipro_fit'].max()
```

```python
# the maximum of carb_fit grouped by label type
print('Max of carb_fit by labels:')
bacteria_data.groupby('labels')['carb_fit'].max()
```

```python
# Count how many rows belong to each Phenotype
print('Count per Phenotype:')
bacteria_data['Phenotype'].value_counts()
```
